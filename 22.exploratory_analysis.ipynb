```python
# Modelando distribuições:
# Distribuição Normal:
sample = np.random.normal(size=1000)
# cria o cdf de sample e plote.

from scipy.stats import norm

xs = np.linspace(-3,3) # cria pontos na reta entre -3 e 3, igualmente espaçados
ys = norm(0,1).cdf(xs) # função normal, encontra os Y's para distribuição normal com média 0 e desvio 1; acdf = acumulada
plt.plot(xs,ys)
```
