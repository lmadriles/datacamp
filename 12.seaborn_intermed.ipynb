# Treliças (trellis, lattice plots ou faceting)

- Exemplos: Facetgrid, factorplot e lmplot
- Faz gráficos lado a lado com mesmas variaveis e escalas nos eixos.
- Requisito: tidy format-> uma observação por linha, variáveis nas colunas

```python
# facetgrid: necessário duas etapas
g = sns.FacetGrid(df, col='HIGHDEG') # seleciona qual coluna discrimina diferentes plots, pode ser row para dispor em linhas
g.map(sns.boxplot, # tipo de gráfico
      'Tuition', # eixo x do boxplot
      col_order=['1','2', '3', '4']) #ordem dos plots de acordo com o valor na coluna 'HIGHDEG'
```

```python
# factorplot: versão rápida e menos flexível
sns.factorplot(x='Tuition', data=df, col='HIGHDEG', kind='box')
```

```python
# facetgrid com regressão (ou scatter): necessário duas etapas
g = sns.FacetGrid(df, col='HIGHDEG') # seleciona qual coluna discrimina diferentes plots
g.map(sns.scatter, # tipo de gráfico
      'Tuition', # eixo x 
      'SAT_AVG_ALL') #eixo y
```

```python
# lmplot
sns.lmplot(data=df, x="Tuition", y="SAT_AVG_ALL",col="HIGHDEG",
           fit_reg=False) # desabilita linhas de regressão
```

### Pairgrid e pairplot

- Reconhecimento de dados = data aware
- Semelhante aos anteriores, a diferença é que definimos apenas as colunas de dados


Paralelo entre duas variáveis. 
Observe os eixos dos plots (a diagonal pricipal tem histograma de cada variável, a secundária mostra scatter entra as duas com eixos alternados)

```python
# pairgrid
g = sns.PairGrid(df, vars=["Fair_Mrkt_Rent","Median_Income"])
g = g.map(plt.scatter) # todos os gráficos scatter

# exemplo
g = sns.PairGrid(df, vars=["Fair_Mrkt_Rent","Median_Income"])
g = g.map_diag(plt.hist)
g = g.map_offdiag(plt.scatter)
```

```python
# mesmo exemplo usando pairplot (atalho do pairgrid)
sns.pairplot(df, vars=["Fair_Mrkt_Rent","Median_Income"], 
             kind='reg', # mostra linha de regressão
             diag_kind='hist')

```

### jointgrid e jointplot

- comparar a distribuição entre duas variáveis
- gráficos de dispersão, linha de regressão, histogramas, distribuição, estimativa de densidade do kernel

Input: variável x e y;

Output: no centro o scatterplt e nos eixos as distribuições

```python
g = sns.JointGrid(data=df, x="Tuition", y="ADM_RATE_ALL") # definindo as duas variáveis
g.plot(sns.regplot, sns.distplot) # definindo primeiro o gráfico central e depois os gráficos dos eixos
```

```python
g = sns.JointGrid(data=df, x="Tuition", y="ADM_RATE_ALL") 

g = g.plot_joint(sns.kdeplot) # definindo tipo do plot do centro
g = g.plot_marginals(sns.kdeplot, shade=True) # definindo tipo do plot dos eixos
g = g.annotate(stats.pearsonr) # informação adicional: R de pearson
```

```python
# jointplot: atalho menos customizável
sns.jointplot(data=df, x="Tuition", y="ADM_RATE_ALL",
              kind='hex') # hexplot: pontos hexagonais; kind must be either 'scatter', 'reg', 'resid', 'kde', or 'hex'
# suporta gráfico de dispersão, hex, residual, regressão e kde
```

```python
g = (sns.jointplot(x="Tuition",
                   y="ADM_RATE_ALL", kind='scatter',
                   xlim=(0, 25000), # valores limites do eixo x
                   marginal_kws=dict(bins=15,rug=True), # customizando os gráficos marginais
                   data=df.query('UG < 2500 & Ownership == "Public"')) # teste lógico p/ slicing
     .plot_joint(sns.kdeplot)) # sobrepõe um gráfico kde sobre o de dispersão (no centro)
```

## Análises de distribuição univariada (uma variável)

distplot() é o melhor pra começar;

rugplot e kdeplot podem ser alternativas úteis.



## Análise de regressão (Entre duas variáveis)
scatter é uma visualização simples pra isso
lmplot() combina em facetgrid
lmplot() comumente a melhor opção em modelos de regressão para determinar relacionamentos lineares entre as variáveis

## Categorical plots

gráficos de estimativa estatística (ponto, barra ou contagem)
