Visualização de dados -  Capítulo 1

```python
import matplotlib.pyplot as plt
```

```python
year = [1950,1970,1990,2010]
pop = [2.519,3.692,5.263,6.972] # população mundial em bilhões
```

```python
plt.plot(year,pop) # plota a linha poligonal
plt.show() #chamar o show permite adicionar customizações
```

```python
plt.scatter(year,pop) # plota apenas os pontos
plt.show() #chamar o show permite adicionar customizações
```

```python
plt.scatter(year,pop)
plt.xscale('log') # poe a escala do eixo x em logaritmica
plt.show() 
```

```python
# Histograma: uma dispersão de dados plotada em x: valor da classe e y: sua frequencia, exemplo: piramide etária
values = [0,0.6,1.4,1.6,2.2,2.5,2.6,3.2,3.5,3.9,4.2,6]
plt.hist(values, bins = 3) # dados, numero de classes (default = 10)
plt.show() # opcional se não houver customização como aqui
```

```python
# plots customizados (ttitulo do gráfico e dos eixos)
plt.plot(year,pop)
plt.xlabel('Year') # eixo x
plt.ylabel('Populatiom') # eixo y
plt.title('Projeção da população mundial') # titulo do grafico
regua = [0,2,4,6,8]
plt.yticks(regua) # define as marcações no eixo y
```

```python
plt.plot(year,pop)
rename_regua = ['0','2B','4B','6B','8B']
plt.yticks(regua,rename_regua) # define e renomeia as marcações no eixo y
plt.show() 
```

```python
# alterando o tamanho da bolinha no scatter
plt.scatter(year,pop, s = [100,200,300,400]) # s = tamanho de cada ponto
```

```python
plt.scatter(year,pop, s = [100,200,300,400], c =  [100,200,300,400], alpha = 0.8) # cor e transparencia (0 a 1)
```

```python
plt.text(1550, 71, 'India') # poe essa palavra nessa coordenada
plt.text(5700, 80, 'China') # poe essa palavra nessa coordenada
plt.grid(True) # desenha a grade no gráfico
```

```python
plt.clf() # limpa a figura / evita que diferentes plots se sobrepunham
```

Capítulo 2 -  Dicionário e pandas
    Funciona igual lista: use os [colchetes] pra adicionar, editar e remover elementos
    Funciona diferente de listas: os indices são as chaves
        se a ordem importa ou se pretende selecionar: use lista; 

```python
dict = {'key': 'value', 'verbete': 'valor'}
print(type(dict))
# tem todos aqueles negocios de dicionario
dict.keys()
dict['verbete'] # acessa o valor dessa chave
# nao aceita lista como chave
dict['verbete'] = 25 # editar
print('key' in dict) #retorna o bool com o resultado do teste
```

```python
del(dict['verbete']) # deleta
dict
```
